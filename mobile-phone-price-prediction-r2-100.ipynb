```python
import numpy as np 
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

# Load Dataset

```python
data=pd.read_csv('/kaggle/input/mobile-phone-price-prediction-cleaned-dataset/Mobile-Price-Prediction-cleaned_data.csv')
```

```python
data.head()
```

```python
data.info()
```

# Features Scaling 

```python
x=data.drop(columns=['Price','Ratings'])
scaler = MinMaxScaler()
scaled_x = scaler.fit_transform(data)
scaled_x
```

```python
y=data['Price']
y
```

# Split Dataset

```python
x_train,x_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.2,train_size=0.8,random_state=0)
```

# Apply Linear Regression Model

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
x.head()
```

# Testing

```python
df=pd.DataFrame({'y_test':y_test,'y_predict':y_predict})
df
```

# Accuracy(R2) 100%

```python
r2Score=r2_score(y_test,y_predict)
r2Score
```
